🎨 Exercício 1: Admissão na Academia de Artes
Este exercício trabalha a precedência de operadores. A condição da audição está vinculada ao treinamento prévio, enquanto o portfólio é um critério independente.

```python
# Entrada de dados
portfolio_forte = input("O candidato possui um portfólio forte? (sim/nao): ").lower()
audicao_excelente = input("A audição foi excelente? (sim/nao): ").lower()
anos_treinamento = int(input("Quantos anos de treinamento prévio o candidato possui? "))

# Lógica de Admissão
# Regra 1: Portfólio Forte (independente)
# Regra 2: Audição Excelente + 2 anos de treino (combinados)
if portfolio_forte == "sim" or (audicao_excelente == "sim" and anos_treinamento >= 2):
    print("Resultado: Candidato Admitido!")
else:
    print("Resultado: Candidato não admitido.")

# EXPLICAÇÃO DOS BLOCOS:
# 1. Entrada: Capturamos o status do portfólio, da audição e o tempo de experiência[cite: 84].
# 2. Processamento: O 'or' separa as duas grandes formas de entrar.
#    O uso de ( ) no 'and' garante que a audição só vale se houver os 2 anos de treino[cite: 87].
# 3. Saída: Informa a decisão final da academia[cite: 88].
```

🏦 Exercício 2: Elegibilidade para Empréstimo Bancário
Neste caso, a renda é um requisito obrigatório (fixo), enquanto o score pode ser atingido pelo próprio cliente ou por um fiador.

```python
# Entrada de dados
renda = float(input("Digite a renda mensal: R$ "))
score_cliente = int(input("Digite o score de crédito do cliente: "))
tem_fiador = input("O cliente possui fiador? (sim/nao): ").lower()

# Inicializamos o score do fiador como 0 caso não exista
score_fiador = 0
if tem_fiador == "sim":
    score_fiador = int(input("Digite o score de crédito do fiador: "))

# Lógica do Empréstimo
# Requisito A: Renda >= 2000 (Obrigatório)
# Requisito B: Score Cliente >= 600 OU Score Fiador >= 700
if renda >= 2000 and (score_cliente >= 600 or score_fiador >= 700):
    print("Resultado: Empréstimo aprovado!")
else:
    print("Resultado: Empréstimo negado.")

# EXPLICAÇÃO DOS BLOCOS:
# 1. Condicional de Fiador: Se não houver fiador, o score dele permanece 0 e não ajuda na conta.
# 2. Operador AND: Garante que sem a renda mínima, nada mais importa[cite: 91].
# 3. Operador OR: Permite flexibilidade no score (próprio ou de terceiro)[cite: 92, 93].
```
