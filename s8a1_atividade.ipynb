Exercício 2: Cálculo de Desconto em Livros

Este programa aplica descontos cumulativos. Primeiro, verifica-se a condição do livro (usando or) e, depois, a quantidade de itens (usando uma verificação independente).




```python
# Entrada de dados
preco_base = float(input("Digite o preço do livro: R$ "))
eh_best_seller = input("O livro é um best-seller? (sim/nao): ").lower()
anos_lancamento = int(input("Há quantos anos o livro foi lançado? "))
quantidade_livros = int(input("Quantos livros está comprando no total? "))

# 1. Verificação do desconto de 20% (Regra: Best-seller OU +2 anos)
# [cite: 17]
if eh_best_seller == "sim" or anos_lancamento > 2:
    desconto_livro = 0.20
    print("Aplicado: 20% de desconto (Regra de categoria/lançamento).")
else:
    desconto_livro = 0.0

# 2. Verificação de desconto adicional de 5% (Regra: > 3 livros)
# [cite: 18]
if quantidade_livros > 3:
    desconto_extra = 0.05
    print("Aplicado: 5% de desconto adicional (Regra de quantidade).")
else:
    desconto_extra = 0.0

# 3. Cálculo Final
# Aplicamos os descontos sobre o preço original [cite: 19]
preco_com_desconto = preco_base * (1 - (desconto_livro + desconto_extra))

print(f"Preço final do livro: R$ {preco_com_desconto:.2f}")

# EXPLICAÇÃO DOS BLOCOS:
# 1. Regra OU (or): Basta uma das condições ser verdadeira para ganhar os 20%.
# 2. Regra Adicional: Uma nova verificação 'if' independente que soma 5% ao desconto total.
# 3. Matemática: Multiplicamos o preço pelo percentual restante para chegar ao valor final.
```
