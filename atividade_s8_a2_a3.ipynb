📘 Atividade S8A2: Doação de Sangue e Cinema
Esta atividade foca em múltiplas condições obrigatórias (and) e exceções combinadas (and / or).

1. Sistema para Doação de Sangue
Neste exercício, utilizamos a lógica onde todas as condições de saúde e segurança precisam ser verdadeiras simultaneamente.

```python
# Entrada de dados
idade = int(input("Digite sua idade: "))
peso = float(input("Digite seu peso (kg): "))
esta_gravida = input("Está grávida? (sim/nao): ").lower()
dias_ultima_doacao = int(input("Há quantos dias foi sua última doação? "))

# Verificação dos requisitos (Uso de AND para condições obrigatórias)
#
if (16 <= idade <= 69) and (peso >= 50) and (esta_gravida == "nao") and (dias_ultima_doacao > 90):
    print("Resultado: Você está apto para doar sangue!")
else:
    print("Resultado: Infelizmente, você não atende aos requisitos para doação.")

# EXPLICAÇÃO:
# - (16 <= idade <= 69): Verifica se a idade está no intervalo permitido[cite: 59].
# - peso >= 50: Garante o peso mínimo necessário[cite: 60].
# - and: Garante que o doador não esteja grávida e respeite o intervalo de dias[cite: 61, 62].
```

2. Controle de Entrada no Cinema
Aqui aplicamos uma regra principal (idade 18+) e uma exceção para menores (15-17 anos) sob certas condições.

```python
# Entrada de dados
idade = int(input("Digite a idade: "))
acompanhado = input("Está acompanhado por um adulto? (sim/nao): ").lower()
tem_autorizacao = input("Possui autorização dos pais? (sim/nao): ").lower()

# Lógica de decisão combinada
#
pode_entrar = False

if idade >= 18:
    pode_entrar = True
elif (15 <= idade <= 17) and (acompanhado == "sim" or tem_autorizacao == "sim"):
    pode_entrar = True

if pode_entrar:
    print("Acesso Permitido!")
else:
    print("Acesso Negado.")

# EXPLICAÇÃO:
# - O primeiro 'if' libera adultos (18+) imediatamente[cite: 65].
# - O 'elif' verifica a faixa de 15 a 17 anos[cite: 67].
# - Dentro do 'elif', usamos 'or' porque basta o acompanhante OU a autorização[cite: 68].
```

📕 Atividade S8A3: Bolsas de Estudo e Vagas de Emprego
Esta atividade explora critérios de elegibilidade onde diferentes caminhos podem levar ao mesmo resultado positivo.

1. Elegibilidade para Bolsa de Estudos
Este programa utiliza o operador or, pois o aluno só precisa atender a um dos critérios de excelência.

```python
# Entrada de dados
gpa = float(input("Digite o GPA do aluno: "))
top_turma = input("O aluno está no top 10% da turma? (sim/nao): ").lower()
horas_voluntario = int(input("Horas de trabalho voluntário no último ano: "))

# Verificação de elegibilidade (Uso de OR para critérios alternativos)
#
if gpa >= 3.5 or top_turma == "sim" or horas_voluntario > 100:
    print("Parabéns! O estudante é elegível para a bolsa de estudos.")
else:
    print("O estudante ainda não atingiu os critérios para a bolsa.")

# EXPLICAÇÃO:
# - Se o GPA for alto (>= 3.5), o aluno já é elegível[cite: 35].
# - Caso contrário, o programa checa se ele é top 10% ou se tem o voluntariado[cite: 36, 37].
```
